```python
import pandas as pd
```

Filtering

```python
#Filtering
# Load the dataset
df = pd.read_csv("students_dataset.csv")
```

```python
df
```

```python
# Filtering:
# Task 1 -> Display the records(rows) of those students whose last name is Gupta
filt = (df['Last_Name'] == 'Gupta')
df.loc[filt]
```

```python
# Task2 -> From the above task print only first_name, Roll_number, and total_marks
df.loc[filt, ['First_Name', 'Roll_Number', 'Total_Marks']]
```

```python
# Task 3 -> Display all other records than the defined filter
df.loc[~filt, ['First_Name', 'Last_Name', 'Roll_Number', 'Total_Marks']]
```

```python
Last_Name_List = ["Gupta", "Thakur"]
filt = df["Last_Name"].isin(Last_Name_List)
```

```python
# Print the records of student whose last name is Gupta or Thakur
df.loc[filt]
```

```python
df.loc[filt, ['First_Name', 'Roll_Number', 'Total_Marks']]
```
