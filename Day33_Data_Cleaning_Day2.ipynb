```python
import pandas as pd
```

```python
df = pd.read_csv("day2_outlier_dataset.csv")
```

Calculate the upper bounds and lower bounds

```python
def iqr_bounds(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return lower, upper
```

```python
# Print above upper bound and lower bound
salary_bounds = df['Salary'].agg(['min', 'max'])
age_bounds = df['Age'].agg(['min', 'max'])

print("Salary bounds:")
print(salary_bounds)
print("\nAge bounds:")
print(age_bounds)
```

Function to filter out the outlier

```python
def remove_outliers(data, column):
    lower, upper = iqr_bounds(data, column)
    filt = ((data[column] >= lower) & (data[column] <= upper))
    return data[filt]
```

Functional Pipeline

```python
df_cleaned =(
      df.pipe(remove_outliers, column = "Salary")
      .pipe(remove_outliers, column = "Age")
)
```
