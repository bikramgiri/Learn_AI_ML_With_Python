```python
import pandas as pd
import matplotlib.pyplot as plt
# Or
# from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv("Data.csv")
data
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
view_count
```

```python
likes
```

```python
ratio
```

```python
plt.scatter(view_count, likes, c = ratio, cmap = "summer", alpha = 0.8, edgecolors = "black", linewidth = 0.7)

cbar = plt.colorbar()
cbar.set_label("Like to Dislike Ratio")

plt.xscale("log")
plt.yscale("log")

plt.xlabel("View Count")
plt.ylabel("Total Likes")
plt.title("Trending YouTube Videos")

plt.tight_layout()
plt.show()
```
