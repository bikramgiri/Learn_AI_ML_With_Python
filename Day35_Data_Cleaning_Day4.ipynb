```python
import pandas as pd
```

```python
file_path = "day4_raw_dataset.csv"
```

```python
df = pd.read_csv(file_path)
df
```

1. Standardize the data(It means Male-M, alice-Alice, and so on)

```python
df["Gender"] = df["Gender"].str.strip().str.capitalize().replace({
      "Male": "M",
      "Female": "F"
})
```

```python
df["Name"] = df["Name"].str.strip().str.capitalize()
```

```python
df["Department"] = df["Department"].str.strip().str.capitalize().replace({
      "Hr": "HR",
      "It": "IT"
})
```

2. Handeling Missing Data

```python
threshold = len(df) * 0.5
```

```python
df = df.dropna(thresh=threshold, axis=1)
```

```python
for col in df.columns:
      if df[col].dtype in ["float64", "int64"]:
            df[col] = df[col].fillna(df[col].median())
      
      else:
            df[col] = df[col].fillna(df[col].mode()[0])
```

3. Convert the joininng data into datetime

```python
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"], errors="coerce")
```

```python
df["Joining_Date"] = df["Joining_Date"].fillna(df["Joining_Date"].mode()[0])
```

4. Ensuring correct datatypes

```python
df["Age"] = df["Age"].astype(int)
df["Salary"] = df["Salary"].astype(int)
df["Performance_Score"] = df["Performance_Score"].astype(int)
```

5. Remove Outlier using IQR

```python
def remove_outliers_IQR(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
```

```python
df = (
      df.pipe(remove_outliers_IQR, column = "Age")
        .pipe(remove_outliers_IQR, column = "Salary")
        .pipe(remove_outliers_IQR, column = "Performance_Score")
)
```

6. Drop the duplicate data

```python
cleaned_data = df.drop_duplicates()
```

7. Save the cleaned data

```python
clean_path = "cleaned_day4_raw_dataset.csv"
df.to_csv(clean_path, index=False)
```
