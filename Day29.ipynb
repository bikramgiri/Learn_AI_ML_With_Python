loc and iloc

```python
# Loc and iloc
# Loc - uses mix of indices and names
df1.loc[[0, 1], ["Hobbyist"]] # Acces one column
```

```python
df1.loc[[0, 1], ["Hobbyist", "OpenSourcer"]] # Access multiple column
```

```python
df1.loc[[0, 1, 2, 3], ["Hobbyist", "OpenSourcer", "Country"]] # Access multiple column
```

```python
# Exception:  ending element pani include garxa loc lae
df1.loc[0:10, "Hobbyist":"Country"] # Access multiple column
```

```python
# iloc : uses only integer indices
# Exception:  ending element pani include garxa loc lae
df1.iloc[[0, 1, 2], [0, 1, 2, 3, 4]]
```

```python
df1.iloc[0:3, 0:5]
```

```python
# Temporarely change indices
df1.set_index("Respondent")
```

```python
# Permanently change indices
df1.set_index("MainBranch", inplace=True)
```
