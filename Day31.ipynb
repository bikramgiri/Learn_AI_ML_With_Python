```python
import pandas as pd
```

```python
data = {
      "A" : [10, 30, 50, 10],
      "B" : [5, 40, 60, 7],
      "C" : [10, 30, 20, 90]
}
```

```python
df = pd.DataFrame(data)
df
```

Basic Arithmetic Operations

```python
# Addition
df["A + B"] = df["A"] + df["B"]
df["A - B"] = df["A"] - df["B"]
df["A * B"] = df["A"] * df["B"]
df["A / B"] = df["A"] / df["B"]
df["A mod B"] = df["A"] % df["B"]
df = df.fillna(0)
df
```

Statistics

```python
df.mean()
```

```python
df.median()
```

```python
df.std()
```

```python
df.min()
```

```python
df.max()
```

Element Wise Operations

```python
import numpy as np
```

```python
df["log_A"] = np.log(df["A"])
df["exp_B"] = np.exp(df["B"])
df["Sq_C"] = np.sqrt(df["C"])
df["Cube_B"] = df["B"].apply(lambda x: x ** 3)
df
```

Applying Functions

```python
def sq(x):
      return x ** 2

df["Square A"] = df["A"].apply(sq)
df
```

```python
def cube(x):
      return x ** 3

df["Cube_B"] = df["B"].apply(cube)
df
```

Column Wise and Row Wise Operations

```python
df["row_sum"] = df.sum(axis=1)
df["row_mean"] = df.mean(axis=1)
df["row_max"] = df.max(axis=1)
df["row_min"] = df.min(axis=1)
df["row_std"] = df.std(axis=1)
df
```

```python
df["col_sum"] = df.sum(axis=0)
df["col_mean"] = df.mean(axis=0)
df["col_max"] = df.max(axis=0)
df["col_min"] = df.min(axis=0)
df["col_std"] = df.std(axis=0)
df
```

```python
col_sum = df.sum(axis=0)
col_sum
```
